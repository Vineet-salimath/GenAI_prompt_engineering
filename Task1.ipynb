Task 1: Replace Hardcoded Prompts

```python

EXPLAIN_TEMPLATE = "Explain {topic} in simple terms for beginners"

explain_prompt = PromptTemplate(
    input_variables=["topic"],
    template=EXPLAIN_TEMPLATE
)

def explain_topic(topic):
    return explain_prompt.format(topic=topic)

print(explain_topic("Machine Learning"))
```

Task 2: Multi-Input Prompt System

```python
MULTI_TEMPLATE = "Explain {topic} for {audience} in a {tone} tone"

multi_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=MULTI_TEMPLATE
)

def build_prompt(topic, audience, tone):
    return multi_prompt.format(topic=topic, audience=audience, tone=tone)

print(build_prompt("AI", "beginners", "friendly"))
print(build_prompt("Python", "kids", "fun"))
print(build_prompt("Deep Learning", "engineers", "technical"))
```

Task 3: Prompt Variations Engine

```python
templates = {
    "teaching": PromptTemplate(
        input_variables=["topic"],
        template="Explain {topic} clearly step by step"
    ),
    "interview": PromptTemplate(
        input_variables=["topic"],
        template="Ask 3 questions about {topic}"
    ),
    "story": PromptTemplate(
        input_variables=["topic"],
        template="Explain {topic} as a story"
    )
}

def generate_variations(topic):
    for key, temp in templates.items():
        print(f"{key.upper()} → {temp.format(topic=topic)}")

generate_variations("Machine Learning")
```

```python
roles = {
    "teacher": "You are a helpful teacher",
    "interviewer": "You are a strict interviewer",
    "motivator": "You are an inspiring mentor"
}

def chat_prompt(role, topic):
    chat = ChatPromptTemplate.from_messages([
        SystemMessagePromptTemplate.from_template(roles[role]),
        HumanMessagePromptTemplate.from_template("Explain {topic}")
    ])
    return chat.format_messages(topic=topic)

msgs = chat_prompt("teacher", "Neural Networks")

for m in msgs:
    print(m.content)
```

Task 5: Input Validation Layer

```python
VALID_AUDIENCE = ["beginner", "intermediate", "expert"]
VALID_TONE = ["formal", "casual", "fun"]

def validate_inputs(audience, tone):
    if audience not in VALID_AUDIENCE:
        audience = "beginner"
    if tone not in VALID_TONE:
        tone = "casual"
    return audience, tone

audience_val, tone_val = validate_inputs("kids", "happy")
print(f"Audience: {audience_val}, Tone: {tone_val}")
```

Task 6: Prompt Generator App

```python
style_templates = {
    "teaching": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} teaching style"
    ),
    "interview": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Give 3 {tone} interview questions on {topic} for {audience}"
    ),
    "story": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} as a {tone} story for {audience}"
    )
}

def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)
    return style_templates[style].format(
        topic=topic,
        audience=audience,
        tone=tone
    )

print(generate_prompt("Neural Networks", "beginner", "fun", "story"))
```


Task 7: Template Reusability Test




```python
reuse_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} to {audience} in a {tone} tone"
)

tests = [
    {"topic": "AI", "audience": "kids", "tone": "fun"},
    {"topic": "Blockchain", "audience": "engineers", "tone": "formal"},
    {"topic": "Python", "audience": "beginner", "tone": "casual"},
    {"topic": "Cybersecurity", "audience": "expert", "tone": "formal"},
    {"topic": "ML", "audience": "student", "tone": "fun"},
]

for t in tests:
    print(reuse_template.format(**t))
```

Pipeline Demo

```python
def run_pipeline(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)
    result = generate_prompt(topic, audience, tone, style)
    print("Final Output:", result)

run_pipeline("Deep Learning", "beginner", "fun", "story")
```
